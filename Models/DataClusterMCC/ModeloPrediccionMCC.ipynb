```python
import csv

with open('datos_generales.txt', 'r') as txtfile:
    txtreader = csv.reader(txtfile, delimiter='|')

    with open('datos_ofuscados.csv', 'w', newline='') as csvfile:
        csvwriter = csv.writer(csvfile)

        for row in txtreader:
            csvwriter.writerow(row)
```

```python
import pandas as pd

file_path = 'datos_ofuscados.csv'
df = pd.read_csv(file_path)

df.head()
```

```python
def get_category(num):
    categories_ranges = {
        "Agricultura": (0, 1499),
        "Servicios_Contratados": (1500, 2999),
        "Transporte": (4000, 4799),
        "Servicios_Utilidad": (4800, 4999),
        "Retail": (5000, 5599),
        "Ropa": (5600, 5699),
        "Miscellaneous": (5700, 7299),
        "Servicios_Negocios": (7300, 7999),
        "Membresias": (8000, 8999),
        "Gobierno": (9000, 9999)
    }
    for category, (start, end) in categories_ranges.items():
        if start <= num <= end:
            return category
    return 'Miscellaneous'
```

```python
df['Category'] = df['MCC'].apply(get_category)
df.to_csv('datos_ofuscados_reprocessed.csv', index=False)
df[['MCC', 'Category']].head()
misc_df = df[df['Category'] == 'Miscellaneous'][['MCC', 'Category']]
print(misc_df)
```
